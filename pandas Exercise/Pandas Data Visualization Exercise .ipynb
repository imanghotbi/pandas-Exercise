# Pandas Data Visualization Exercise

This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r'df3.csv')
%matplotlib inline
```

```python
data.info()
```

```python
data.head()
```

** Recreate this scatter plot of b vs a. Note the color and size of the points.**

```python
data.plot.scatter(x="a",y="b",s=50,figsize=(13,5),color="red")
```

** Create a histogram of the 'a' column.**

```python
data["a"].plot.hist()
```

** These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
sns.set_style("darkgrid")
data['a'].plot.hist(bins=20,color="#2ed22d",alpha=0.5)
plt.xlim([0,1])
```

** Create a boxplot comparing the a and b columns.**

```python
data.boxplot(column=['a','b'])
plt.show()
```

** Create a kde plot of the 'd' column **

```python
data["d"].plot.kde(linestyle='dashed',color='red',linewidth=4)
plt.xlim([-0.5,1.5])
plt.ylim([0,1.2])
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

** Create an area plot of all the columns for just the rows up to 30.**

```python
data[:31].plot.area(alpha=0.5)
plt.xlim([0,30])
plt.legend(loc='center right',bbox_to_anchor=(1.15,.5))
```
